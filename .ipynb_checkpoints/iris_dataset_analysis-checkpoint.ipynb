# Train test split
You always want to evaluateyour final model on a test set that hasn't been used at all in the training process. So we'll split off a test set here.

_(Note: This might be a little less true if you use cross-validation, but is still considered a best practice.)_

````python
from sklearn.model_selection import train_test_split
````

````python
df_train, df_test = train_test_split(df, test_size=0.25)
````

````python
df_train.shape
````

````python
df_test.shape
````

````python
df_train.head()
````

# Prepare our data for modeling
This involves splitting the dta back out into plain NumPy arrays

````python
X_train = df_train.drop(columns=["target", "target_name"]).values
y_train = df_train["target"].values
````

````python
print("X:", X_train[:5])
print("y:", y_train[:5])
````

# Modeling - What is our baseline?
What is the simplest model we can think of?

In this case, if our baseline omdel is just randomly guessing the species of flower, or guessing a singel species for every data point, we would expect to have model accuracy of 0.33 or 33%, since we have 3 different classes that are evenly balanced.

So our models shohuld at least beat 33% accuracy.

# Modeling - Seimple manual model
Let's manyally look at a our data and decide some cutoff points for classification.

````python
def single_feature_prediction(petal_length):
    """Predicts the Iris species given the petal length."""
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
````

````python
df_train.columns
````

````python
X_train[:, 2]
````

````python
# Python list comprehension
manual_y_predictions = np.array([single_feature_prediction(val) for val in X_train[:, 2]])
````

````python
manual_y_predictions
````

````python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
````

````python
print(f"Manual model accuracy: {manual_model_accuracy * 100:.2f}%")
````

# Using cross-validation to validate model

````python
accuracies = cross_val_score(estimator=model, X=X_train, y=y_train, cv=5, scoring="accuracy")
````

````python
sns.scatterplot(x="petal length (cm)", y="petal width (cm)", hue="target_name", data=df_predictions)
````

````python
def plot_incorrect_predictions(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots (2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="prediction_label", data=df_predictions, ax=axs[0])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="target_name", data=df_predictions, ax=axs[1])
    sns.scatterplot(x=x_axis_feature, y=y_axis_feature, hue="correct_prediction", data=df_predictions, ax=axs[2])
    axs[3].set_visible(False)
    
    plt.show()
````

````python
plot_incorrect_predictions(df_predictions, "petal length (cm)", "petal width (cm)")
````

# Model Tuning
What is model tuning?

Model tuing is trying to determine the parameters of your model (these are the also known as "hyperparameters") that maximize the model performance

````python
# model = RandomForestClassifier()
````

````python
for reg_param in (0.1, 1, 2, 6, 8):
    print(reg_param)
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, X_train, y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100:.2f}%")
````

# Final Model

````python
model = LogisticRegression(max_iter=200, C=3)
````
